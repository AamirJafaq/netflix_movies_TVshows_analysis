```python
import pandas as pd
```

```python
data=pd.read_csv("netflix_titles.csv")
data.head()
```

The dataset has 8807 observations.

```python
data.shape
```

Getting columns with atleast one missing value.

```python
missing_col=data.isna().sum()
missing_col=missing_col[missing_col>0]
missing_col
```

```python
data.rename(columns={"cast": "casts"}, inplace=True)
```

Note that we can not fill the director name randomly. So, for simplicity we will put unknow where directors names are not given. On the similar notes, we will replace the null values in cast and country as unknown.

```python
data["director"].fillna("unknow", inplace=True)
data["country"].fillna("unknow", inplace=True)
data["casts"].fillna("unknown", inplace=True)
```

Now, we will replace the null values in date_added column with the mode value. date_added is date when movies was added on netflix. It does makes sense to replace null values with mode value.

```python
data['date_added'].fillna(data["date_added"].mode()[0], inplace=True)
```

In other columns rating and duration, there are 3 to 4 values are null. So, it can be drop or replace with mode values. 

```python
data.dropna(subset=["rating", "duration"], inplace=True)
```

```python
data.shape
```

The final dataset has 8800 observations.

Save the final data as csv file.

```python
data.to_csv("CLEAN_netflix_titles.csv", index=False)
```
